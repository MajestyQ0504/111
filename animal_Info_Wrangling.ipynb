## Importing libraries

```python
# Importing required libraries for wrangling
import pandas as pd
pd.set_option('display.max_columns', 500)
import numpy as np
```

### Reading the Animal Information file

```python
# reading the animal information file
animal_Info = pd.read_csv('records-2020-04-09.csv')  
```

```python
# checking the format of the file
animal_Info.head()
```

```python
# getting the list of columns
list(animal_Info.columns)
```

```python
# Remobing inwanted columns
animal_Info = animal_Info.drop(columns=['Record ID', 'Catalogue Number','Taxon Concept GUID','Phylum','Taxon Rank','Kingdom','Subspecies','Data Resource ID','Data Resource Name','Institution ID','Institution','Collection ID','Collection','Licence','Institution Code','Collection Code','Locality','Coordinate Precision','IMCRA 4 Regions','Minimum elevation in meters','Maximum elevation in meters','Minimum depth in meters','Individual count','Basis Of Record - original','Basis Of Record','Occurrence status','Sex','Preparations','Outlier for layer','Taxonomic Quality','Location Quality','Maximum depth in meters','Year','Month','Day','Event Date - parsed','Verbatim event date','Vernacular name - original','Scientific Name - original'])
```

## Checking for Missing Values:

```python
animal_Info.info()
```

```python
# Replacing missing values with Nan
animal_Info['Vernacular name'].replace('', np.nan, inplace=True)
```

```python
# dropping rows with Nan
animal_Info.dropna(subset=['Vernacular name'], inplace=True)
```

```python
animal_Info.info()
```

```python
# Filling the missing distances with average radius
radius_avg = animal_Info['Coordinate Uncertainty in Metres'].mean()
distance_radius = radius_avg.round(0)
distance_radius
```

```python
animal_Info['Coordinate Uncertainty in Metres'].fillna(2566, inplace=True)
```

```python
animal_Info.info()
```

### writing wrangled data to new csv file

```python
animal_Info.to_csv('Animal_Info.csv')
```
